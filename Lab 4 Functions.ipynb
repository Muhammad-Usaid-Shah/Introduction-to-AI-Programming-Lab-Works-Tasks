OHM LAW CALCULATER

```python
def ohms_law(I, R=1):
    V = I * R
    return V

# Positional arguments
print("Positional:", ohms_law(2, 5))

# Keyword arguments
print("Keyword:", ohms_law(I=3, R=4))

# Default argument (only I given, R=1 by default)
print("Default:", ohms_law(10))

# Mixing positional and keyword
print("Mixing:", ohms_law(2, R=6))


```

3. Solar Battery Charging:

```python
def charge_from_solar(SOC: float) -> None:
    """
    This function checks the battery's State of Charge (SOC) 
    and prints the charging status.

    Parameters:
    SOC (float): The current state of charge in percentage (0â€“100).
    
    Returns:
    None
    """
    
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print(f"Battery charging, SOC = {SOC}%")

# Printing the docstring of the function
print(charge_from_solar.__doc__)
```

Returning as a Tuple

MAIN PROGRAM

resistance in ohm


```python
def resistance():
    return [10, 20, 30, 40, 50] 
```

voltage in volts

```python
def voltage():
    return [5, 10, 15, 20, 25]  
```

calculate current in ohm law

```python
def current(res_list, volt_list):
    currents = []
    for i in range(len(res_list)):
        I = volt_list[i] / res_list[i]   # Ohm's Law
        currents.append(I)
    return (res_list, volt_list, currents)
```

main program

```python
res, volt, curr = current(resistance(), voltage())  # unpack tuple

print("Resistances:", res)
print("Voltages:", volt)
print("Currents:", curr)
```

ðŸ“Œ Version 2: Returning as a Dictionary

return dictionary

```python
def current_dict(res_list, volt_list):
    currents = {}
    for i in range(len(res_list)):
        I = volt_list[i] / res_list[i]
        currents[f"current_{i+1}"] = I   # key like current_1, current_2
    return currents
```

main program

```python
res_list = resistance()
volt_list = voltage()
curr_dict = current_dict(res_list, volt_list)

print("Resistances:", res_list)
print("Voltages:", volt_list)
print("Currents:", curr_dict)
```

```python
def series(resistances: list | tuple) -> float:
    """
    This function calculates the total resistance 
    of resistors connected in series.

    Parameters:
    resistances (list | tuple): A collection of resistor values in ohms.

    Returns:
    float: The total resistance in ohms.
    """
    return sum(resistances)

# Example usage
res_list = [10, 20, 30, 40, 50]
total_R = series(res_list)

print("Resistances:", res_list)
print("Total Series Resistance:", total_R)

# Print function annotations
print(series.__annotations__)
```
