Lab 11: Dictionary Comprehension

Task 1-Map Device Names to Power Ratings

```python
# List of device names

devices = ["Fan", "Light", "AC", "Pump"]
```

```python
# Corresponding power ratings in watts

power = [60, 15, 1800, 750]
```

```python
# Create dictionary using dictionary comprehension

device_power = {devices[i]: power[i] for i in range(len(devices))}
```

```python
# Print the dictionary

print(device_power)  
```

Task 2-Categorize Resistances as Low or High

```python
# List of resistances in ohms

R = [5, 33, 47, 82, 10]
```

```python
# Create dictionary using comprehension

resistance_category = {value: ("Low" if value < 40 else "High") for value in R}
```

```python
# Print the dictionary

print(resistance_category)  
```

Task 3-Compute Current for Each Load (I = P/V)

```python
# Loads in watts

loads = [60, 100, 150, 200]
```

```python
# Supply voltage in volts

voltage = 200
```

```python
# Create dictionary using comprehension

current_dict = {load: load / voltage for load in loads}
```

```python
# Print the dictionary

print(current_dict)  
```

Task 4: Convert Dictionary Values From °C to °F

```python
#Dictionary

temps_c = {"motor": 40, "transformer": 60, "room": 28}
```

```python
# convert Celsius to Fahrenheit

temps_f = {k: (v * 9/5) + 32 for k, v in temps_c.items()}
```

```python
print(temps_f)
```

Task 5: Create Dict of Components With Index Keys

```python
#List Components

components = ["Resistor", "Capacitor", "Inductor", "Diode"]
```

```python
# create dictionary using index as key

components_dict = {i: name for i, name in enumerate(components)}
```

```python
print(components_dict)
```

Task 6: Filter Only Devices With Power > 100W

```python
#Make a dictionary

device_power = {
    "Fan": 60,
    "Light": 15,
    "Heater": 1200,
    "Iron": 1000,
    "Router": 10
}
```

```python
# keep devices with power greater than 100W

high_power_devices = {k: v for k, v in device_power.items() if v > 100}
```

```python
print(high_power_devices)
```
