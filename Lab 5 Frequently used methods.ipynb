```python
#Lab-5
#1. enumerate()
#i. Voltage readings
voltages = [12.5, 19.2, 24.0, 11.7, 30.1]
for i, v in enumerate(voltages):
    print(i, v)
```

```python
#ii. Resistor values
resistors = [100, 220, 330, 470, 1000]
for i, r in enumerate(resistors):
    print(i, r)
```

```python
#2. zip()
#i. Current + Voltage pairs
currents = [2, 3, 1.5, 4]
voltages = [12, 24, 9, 5]
for c, v in zip(currents, voltages):
    print(c, v)
```

```python
#ii. Components + Ratings
components = ["Resistor", "Capacitor", "Inductor"]
ratings = ["220 ohm", "10 µF", "5 mH"]
for c, r in zip(components, ratings):
    print(c, r)
```

```python
#3. map()
#i. Celsius → Fahrenheit
C = [0, 25, 40]
F = list(map(lambda x: (x*9/5) + 32, C))
print(F)
```

```python
#ii. Volts → Kilovolts
V = [1000, 2500, 500]
kV = list(map(lambda x: x/1000, V))
print(kV)
```

```python
#4. filter()
#i. Currents > 10 A
curr_readings = [5, 12, 18, 7, 25]
high_curr = list(filter(lambda x: x > 10, curr_readings))
print(high_curr)
```

```python
#ii. Capacitors > 50 µF
caps = [10, 20, 50, 100]
big_caps = list(filter(lambda x: x > 50, caps))
print(big_caps)
```

```python
#5. sorted()
#i. Resistance increasing
res = [330, 100, 220, 470]
print(sorted(res))
```

```python
#ii. Marks descending
marks = [55, 89, 72, 91]
print(sorted(marks, reverse=True))
```

```python
#6. reversed()
#i. Power consumption reversed
power = [100, 150, 200, 180]
print(list(reversed(power)))
```

```python
#ii. Current measurements reversed
curr_list = [5, 10, 15, 20]
print(list(reversed(curr_list)))
```

```python
#7. range()
#i. Time slots 0–10 sec
for t in range(0, 11):
    print(t)
```

```python
#ii. Resistor IDs 1–20
for r in range(1, 21):
    print(r)
```

```python
#8. any()
#i. Any voltage > 220 V
vlist = [110, 150, 230, 180]
print(any(x > 220 for x in vlist))
```

```python
#ii. Any motor overloaded
motor_load = [False, False, True, False]
print(any(motor_load))
```

```python
#9. all()
#i. All resistors > 100 ohm
res_list = [150, 200, 330]
print(all(x > 100 for x in res_list))
```

```python
#ii. All solar panels active
panels = [True, True, True]
print(all(panels))
```

```python
#10. isinstance()
#i. Check if current is integer
current = 5
print(isinstance(current, int))
```

```python
#ii. Check rating is string or number
rating = "220 ohm"
print(isinstance(rating, str))
```

```python
#11. type()
#i. Type of resistance (float)
res_value = 47.5
print(type(res_value))
```

```python
#ii. Type of voltage list
voltage_list = [12, 24, 36]
print(type(voltage_list))
```

```python
#12. len()
#i. Count resistors
resistors = [100, 220, 330, 470]
print(len(resistors))
```

```python
#ii. Characters in project name
project_name = "Power Analyzer"
print(len(project_name))
```

```python
#13. sum()
#i. Total power
powers = [10, 20, 30, 40]
print(sum(powers))
```

```python
#ii. Series resistance
series_res = [10, 20, 30]
print(sum(series_res))
```

```python
#14. max()
#i. Highest voltage
v = [110, 150, 90, 200]
print(max(v))
```

```python
#ii. Maximum motor current
currs = [5, 9, 12, 7]
print(max(currs))
```

```python
#15. min()
#i. Lowest temperature
temps = [20, 35, 15, 40]
print(min(temps))
```

```python
#ii. Smallest resistor
rl = [100, 10, 220, 330]
print(min(rl))
```

```python
#16. abs()
#i. Absolute AC readings
ac = [-5, 10, -15, 20]
print(list(map(abs, ac)))
```

```python
#ii. Magnitude of negative voltage
v = -12
print(abs(v))
```

```python
#17. round()
#i. Round voltage
vol = 12.6789
print(round(vol, 2))
```

```python
#ii. Round frequency
freq = 50.8234
print(round(freq))
```

```python
#18. divmod()
#i. Fiber ducts
print(divmod(87, 10))
```

```python
#ii. LEDs groups
print(divmod(42, 12))
```

```python
#19. pow()
#i. Square of current values
currs = [2, 3, 4]
print([pow(i, 2) for i in currs])
```
