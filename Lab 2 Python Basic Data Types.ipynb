4. Bit Parity Checker

```python
# Input binary signal as integer
signal = 0b11010110

# Count number of 1s
ones_count = signal.bit_count()

# Determine parity
if ones_count % 2 == 0:
    print(f"Binary signal {bin(signal)} has EVEN parity ({ones_count} ones).")
else:
    print(f"Binary signal {bin(signal)} has ODD parity ({ones_count} ones).")
```

```python
# Compute angular frequency
w = 2 * math.pi * f
```

```python
# Impedances
Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))
```

```python
print("Capacitor impedence(Z_R):",Z_R)
```

```python
# Print results", Z_R)
print("Inductor Impedance (Z_L):", Z_L)
print("Capacitor Impedance (Z_C):", Z_C)
```

Task 6: Find Total Impedance

```python
import cmath
import math

z1=2+3j
z2=3-6j
z3=4+4j

ztotal=z1+z2+z3
```

```python
print("Imaginary",ztotal.imag,"ohms")
print("Real",ztotal.real,"ohms")
```

```python
magnitude=abs(ztotal)
phase_angle=cmath.phase(ztotal)
```

```python
print("Magnitude",magnitude,"ohms")
print("Phase angle",phase_angle,"ohms")
```

```python
#Ohms law
V = 230 
I = 5 
R = V / I 

# Using f-string
print(f"""The voltage is {V} V, the current is {I} A, so the resistance 
is {R:.2f} Ω.""")

# Using .format()
print("""The voltage is {} V, the current is {} A, so the resistance is 
{:.2f} Ω.""".format(V, I, R))
```

```python
#AC Power
P = 2000 # watts
V = 220 # volts
I = P / V # amps
```

```python
print(f"The real power is {P} W at {V} V supply, drawing {I:.2f} A current.")
print("The real power is {0} W at {1} V supply, drawing {2:.2f} A current.".format(P, V, I))
```

```python
#RLC Circuit Impedance
R = 50
X_L = 30
X_C = 20
Z = complex(R, X_L - X_C)
print(f"Impedance of the circuit is {Z.real} + j{Z.imag} Ω.")
print("Impedance of the circuit is {0} + j{1} Ω.".format(Z.real,Z.imag))
```

```python
#Efficiency calculation
P_out = 950 # watts
P_in = 1000 # watts
eff = (P_out / P_in) * 100
print(f"""The machine efficiency is {eff:.1f}% when output is {P_out} W 
and input is {P_in} W.""")
print("""The machine efficiency is {:.1f}% when output is {} W and input 
is {} W.""".format(eff, P_out, P_in))
```
