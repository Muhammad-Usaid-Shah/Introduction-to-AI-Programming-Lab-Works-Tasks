Task 1-Basic List Comprehension

```python
# Create a list of numbers from 0 to 9

numbers = [i for i in range(10)]  
```

Task 2-Filtering Items with if

```python
# Create a List

nums = [1, 2, 3, 4, 5, 6]
```

```python
# Keep only even numbers 

even_nums = [n for n in nums if n % 2 == 0]  
```

```python
# Print result

print(even_nums)  
```

Task 3-Conditional Expression Inside Comprehension

```python
# Create a list

nums = [1, 2, 3, 4]
```

```python
# Make a list where even numbers become "even" and odd become "odd"

labels = [f"{n} is even" if n % 2 == 0 else f"{n} is odd" for n in nums]
```

```python
# Print result

print(labels)
```

Task 1: Filter Voltage Readings

```python
#Voltages List

voltages = [224, 198, 231, 177, 205, 189, 240]
```

```python
# get voltages below 200

low_voltages = [v for v in voltages if v < 200]
```

```python
print(low_voltages)
```

Task 2: Convert Power Ratings to kW

```python
#Power List

powers_w = [50, 120, 300, 750, 1500]
```

```python
# convert watts to kilowatts

powers_kw = [p / 1000 for p in powers_w]
```

```python
print(powers_kw)
```

Task 3: Square Only Even Resistances

```python
#Resistances List

resistors = [10, 15, 22, 47, 50, 68]
```

```python
# square only even values

even_squares = [r**2 for r in resistors if r % 2 == 0]
```

```python
print(even_squares)
```

Task 4: Label Currents as "High" or "Low"

```python
#Currents List

currents = [2.1, 0.5, 1.8, 3.0, 0.9]
```

```python
# label current values

current_labels = ["High" if c > 1.5 else "Low" for c in currents]
```

```python
print(current_labels)
```

Task 5: Extract Appliance Names Starting With ‘F’

```python
#Appliances List

appliances = ["Fan", "Fridge", "Light", "Freezer", "Pump"]
```

```python
# get names starting with F

f_appliances = [a for a in appliances if a.startswith("F")]
```

```python
print(f_appliances)
```
