Task 1 — Smart Home Energy Meter

```python
meter = {
    "time": "18:32",
    "grid_power": 720,
    "solar_power": 340,
    "battery_soc": 67,
    "appliance_status": {"AC": "OFF", "Fridge": "ON", "Motor": "OFF"}
}
```

```python
#1. Print the "solar_power" value

print(meter["solar_power"])
```

```python
#2. Add new appliance "TV": "ON"

meter["appliance_status"]["TV"] = "ON"
```

```python
#3. Change "battery_soc" to 80

meter["battery_soc"] = 80
```

```python
#4. Check if "solar_power" exists

print("solar_power" in meter)
```

```python
#5. Delete "Motor" from appliances

del meter["appliance_status"]["Motor"]
```

Task 3 — Weather Monitoring Station

```python
weather = {
    "hour": 14,
    "temperature": 33.8,
    "humidity": 52,
    "wind": {"speed": 12, "direction": "NE"},
    "alerts": []
}
```

```python
#1. Print wind direction

print(weather["wind"]["direction"])
```

Task 4-Inventory System for Electronics Shop

```python
# Inventory dictionary
inventory = {
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {
        "voltage": "12/24V",
        "current": "20A",
        "brand": "SolarTech"
    }
}
```

```python
# 1. Print the brand name

print(inventory["features"]["brand"])  
```

```python
# 2. Reduce stock by 2

inventory["stock"] -= 2
print(inventory["stock"]) 
```

```python
# 3. Add warranty feature

inventory["features"]["warranty"] = "1 year"
print(inventory["features"])  
```

```python
# 4. Check if price key exists

print("price" in inventory) 
```

```python
# 5. Add discount to inventory

inventory.update({"discount": 10})
print(inventory) 
```

Task 5-Hospital Patient Monitoring System

```python
# Patient dictionary
patient = {
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {
        "aspirin": "8AM",
        "vitamin_D": "12PM"
    }
}
```

```python
# 1. Print oxygen level

print(patient["oxygen_level"])  
```

```python
# 2. Add new medication

patient["medications"]["antibiotic"] = "6PM"
print(patient["medications"])  
```

```python
# 3. Update heart rate

patient["heart_rate"] = 78
print(patient["heart_rate"])  
```

```python
# 4. Get insulin schedule safely

print(patient["medications"].get("insulin", "Not Prescribed"))  
```

```python
# 5. Print all key-value pairs

for key, value in patient.items():
    print(key, ":", value)  
```
