```python
#Lab Report Task
#Three Impedance
import cmath
Z1=2+3j
Z2=3-1j
Z3=2+4j
#Total Impedance
Z=Z1+Z2+Z3
#Real(Resistive) and Imaginary(Reactive) Part
print("Real part:",Z.real)
print("Imaginary part:",Z.imag)
#Magnitude and Phase angle
print("Magnitude:",abs(Z))
print("Phase angle(radians):",cmath.phase(Z))
print("Phase angle(Degrees):",cmath.phase(Z)*180/3.14)
```

```python
#Task 5:
import cmath
import math

# Circuit parameters
R = 50      # ohms
L = 0.1     # henry
C = 100e-6  # farad
f = 60      # hertz
```

```python
# Impedances
Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))
```

```python
print("Capacitor impedence(Z_R):",Z_R)
```

```python
# Print results", Z_R)
print("Inductor Impedance (Z_L):", Z_L)
print("Capacitor Impedance (Z_C):", Z_C)
```

```python
#c. RLC Circuit Impedance
R = 50
X_L = 30
X_C = 20
Z = complex(R, X_L - X_C)
print(f"Impedance of the circuit is {Z.real} + j{Z.imag} Ω.")
print("Impedance of the circuit is {0} + j{1} Ω.".format(Z.real,Z.imag))
```
