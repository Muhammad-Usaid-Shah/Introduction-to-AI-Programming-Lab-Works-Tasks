```python
with open("volatage.txt", "r", encoding="latin-1") as f:
    voltages = f.readlines()

with open("voltage_results.txt", "w", encoding="utf-8") as f:
    for v in voltages:
        v = v.strip()       
        v = int(v)
        if v < 110:
            f.write("Low\n")
        elif v <= 220:
            f.write("Medium\n")
        else:
            f.write("High\n")

print("Voltage classification saved to voltage_results.txt")
```

```python
# Task: 3
with open("scores.txt", "r", encoding="latin-1") as f:
    scores = f.readlines()

with open("test_report.txt", "w") as f:
    for s in scores:
        s = s.strip()
       
        if s.isdigit():
            s = int(s)
            if s >= 50:
                f.write("Pass\n")
            else:
                f.write("Fail\n")
        else:
            f.write("Invalid\n")

print("Pass/Fail results saved to test_report.txt")
```

task 4

```python
# Task: 4 Pass/Fail Test Report (Engineering Labs)
# Input: scores.txt (each line contains one student's lab test score)
# Output: results.txt (each line contains either "Pass" or "Fail")

with open("scores.txt", "r", encoding="latin-1") as input_file:
    score_lines = input_file.readlines()

# Step 2: Evaluate each score and write result
with open("results.txt", "w") as output_file:
    for line in score_lines:
        line = line.strip()
        if not line:  # skip empty lines
            continue
        try:
            score = int(line)  # Try converting to integer
            if score >= 50:
                output_file.write("Pass\n")
            else:
                output_file.write("Fail\n")
        except ValueError:
            output_file.write("Invalid\n")  # Handle non-numeric entries

print("✅ Test report saved to results.txt")
```

task 5

```python
# Task 5: File Extension Validator
import os

# Task: Validate file extensions in the 'reports' folder
path = "reports"  # Folder containing files

# Check if folder exists
if not os.path.exists(path):
    print("❌ Folder 'reports' not found.")
else:
    files = os.listdir(path)

    pdf_count = docx_count = invalid_count = 0

    for file in files:
        ext = file.lower()  # Make extension check case-insensitive
        if ext.endswith(".pdf"):
            print(file, "→ Valid Report")
            pdf_count += 1
        elif ext.endswith(".docx"):
            print(file, "→ Needs Conversion")
            docx_count += 1
        else:
            print(file, "→ Invalid Format")
            invalid_count += 1

    # Summary
    print("\nSummary:")
    print("PDF =", pdf_count)
    print("DOCX =", docx_count)
    print("Invalid =", invalid_count)
```
