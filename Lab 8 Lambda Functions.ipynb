```python
# Task 1: Basic Lambda
square = lambda x: x * x
print(f"Input 4: {square(4)}")
print(f"Input -3: {square(-3)}")
```

```python
# Task:2 Lambda With Multiple Arguments

# Two Inputs (Addition)
add = lambda a, b: a + b
print(f"Two inputs (2 + 5) = {add(2, 5)}")
```

```python
# Three Inputs (Volume)
volume = lambda l, w, h: l * w * h
print(f"Three inputs (2, 3, 4) = {volume(2, 3, 4)}")
```

```python
# Keyword-Only Argument (Simulated with Default Value)
add_offset = lambda x, offset=10: x + offset
print(f"Keyword-simulated (5 + default): {add_offset(5)}")
```

```python
# Task:3 Lambda With map()

numbers = [1, 2, 3, 4]
strings = ["resistor", "inductor", "cap"]
tuples = [('A', 10), ('B', 5), ('C', 20)]
```

```python
# Transform List of Numbers (Cube)
cubed = list(map(lambda x: x**3, numbers))
print(f"Map Numbers (Cubed): {cubed}")
```

```python
# Transform List of Strings (Uppercase)
uppercase = list(map(lambda s: s.upper(), strings))
print(f"Map Strings (Uppercase): {uppercase}")
```

```python
# Transform List of Tuples (Get first element and square the second)
transformed_tuples = list(map(lambda t: (t[0], t[1]**2), tuples))
print(f"Map Tuples (Squared value): {transformed_tuples}")
```

```python
# Task:4 Lambda With filter()

data = [1, 8, 3, 10, 5, 12]
words = ["A", "Wire", "Bus", "Chip", "V"]
mixed = [10, "Component", True, 4.5, "R"]
```

```python
# Keep Only Even Numbers
evens = list(filter(lambda x: x % 2 == 0, data))
print(f"Filter Evens: {evens}")
```

```python
# Keep Only Strings Longer Than 3 Characters
long_strings = list(filter(lambda s: len(s) > 3, words))
print(f"Filter Long Strings: {long_strings}")
```

```python
# Keep Only Integers from Mixed List
ints_only = list(filter(lambda x: isinstance(x, int), mixed))
print(f"Filter Integers: {ints_only}")
```

```python
# Task:5 Lambda With sorted() key

words_list = ["IC", "Diode", "Transistor", "LED"]
data_tuples = [('R1', 100), ('R2', 50), ('R3', 220)]
power_ratings = {'LED': 5, 'Fan': 60, 'Motor': 350, 'Bulb': 40}
```

```python
# Sort List by Length
by_length = sorted(words_list, key=lambda s: len(s))
print(f"Sort by Length: {by_length}")
```

```python
# Sort List of Tuples by Second Element
by_value = sorted(data_tuples, key=lambda t: t[1])
print(f"Sort by Second Element: {by_value}")
```

```python
# Sort Dictionary Items by Their Values
sorted_items = sorted(power_ratings.items(), key=lambda item: item[1])
print(f"Sort Dict by Value: {sorted_items}")
```

```python
# Task:6 Lambda With reduce()

from functools import reduce
numbers = [1, 2, 3, 4]
strings = ["Data", "Bus", "Line"]
```

```python
# Combine All Numbers (Summation)
sum_all = reduce(lambda x, y: x + y, numbers)
print(f"Reduce Sum: {sum_all}")
```

```python
# Join Multiple Strings (Concatenation)
joined_string = reduce(lambda s1, s2: s1 + " " + s2, strings)
print(f"Reduce Join: {joined_string}")
```

```python
# Find Custom Largest Element
largest = reduce(lambda x, y: x if x > y else y, numbers)
print(f"Reduce Largest: {largest}")
```

```python
# Task:7 Replace a small def with a lambda and compare readability

def label_meter(id):
    return "Meter-" + str(id).zfill(3)
```

```python
label_meter_lambda = lambda id: "Meter-" + str(id).zfill(3)
```

```python
print(f"Def version (12): {label_meter(12)}")
print(f"Lambda version (12): {label_meter_lambda(12)}")
```

```python
# Task:8 Sort electrical components using lambda key

import operator
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]
```

```python
# Sorting with lambda
sorted_lambda = sorted(components, key=lambda item: item[1])
print(f"Sorted (Lambda): {sorted_lambda}")
```

```python
# Sorting with operator.itemgetter
sorted_itemgetter = sorted(components, key=operator.itemgetter(1))
print(f"Sorted (itemgetter): {sorted_itemgetter}")
```

```python
# Task:9 Convert sensor strings using map() + lambda

readings = ["2.5", "5.0", "3.3"]
CALIBRATION_OFFSET = 0.1
```

```python
# map() + lambda Version
calibrated_map = list(map(lambda s: float(s) + CALIBRATION_OFFSET, readings))
print(f"Map + Lambda: {calibrated_map}")
```

```python
# List Comprehension Equivalent
calibrated_list_comp = [float(s) + CALIBRATION_OFFSET for s in readings]
print(f"List Comprehension: {calibrated_list_comp}")
```

```python
# Task:10 Filter out low power values using filter() + lambda

powers = [20, 75, 45, 120, 10]
THRESHOLD = 50
```

```python
# filter() Version
high_power_filter = list(filter(lambda p: p > THRESHOLD, powers))
print(f"Filter + Lambda: {high_power_filter}")
```

```python
# List Comprehension Version
high_power_list_comp = [p for p in powers if p > THRESHOLD]
print(f"List Comprehension: {high_power_list_comp}")
```

```python
# Task:11 Use reduce() to combine values (total impedance calculation)

from functools import reduce
import math
Z = [2, 3, 4] 
```

```python
# reduce() + lambda (Product)
total_impedance_reduce = reduce(lambda x, y: x * y, Z)
print(f"Reduce Product: {total_impedance_reduce} \u03A9")
```

```python
# Simpler Alternative (math.prod)
total_impedance_prod = math.prod(Z)
print(f"Math.prod: {total_impedance_prod} \u03A9")
```

```python
# Task:12 Conditional lambda for load classification

# Ternary operator: (True_Value if Condition else False_Value)
classify_load = lambda power: "High Load" if power > 100 else "Low/Normal Load"
```

```python
print(f"150 W: {classify_load(150)}")
print(f"90 W: {classify_load(90)}")
```
