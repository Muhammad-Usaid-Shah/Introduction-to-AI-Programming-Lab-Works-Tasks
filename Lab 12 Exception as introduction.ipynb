Task 1: Type Error (Add int and string)

```python
#Step 1: Run This Code

a = 10
b = "5"
result = a + b
print(result)
```

```python
#Step 2: Fix the Code

a = 10
b = "5"

try:
    result = a + b
    print(result)
    
except TypeError:  
    print("Error: Integer and a String cannot be added.")
```

Task 2: Division by Zero

```python
#Step 1: Run This Code

x = 20
y = 0
result = x / y
print(result)
```

```python
#Step 2: Fix the Code

x = 20
y = 0

try:
    result = x / y
    print("Result:", result)
    
except ZeroDivisionError:  
    print("Error: Division by zero is not allowed.")
```

Task 3: List Index Out of Range

```python
#Step 1: Run This Code

numbers = [10, 20, 30]
print(numbers[5])
```

```python
#Step 2: Fix the Code

numbers = [10, 20, 30]

try:
    print(numbers[5])
    
except IndexError:  
    print("Error: Invalid index. Please use an index within the list range.")
```

Task 4: Logical Error Using raise

```python
#Step 1: Run This Code

age = -5
print("Age:", age)
```

```python
#Step 2: Fix the Code

age = -5

try:
    if age < 0:
        raise Exception("Error: Age cannot be negative.")

    print("Age:", age)

except Exception as e:  
    print(e)
```

Task 5: Assert Failure

```python
#Step 1: Run This Code

total_marks = 0
assert total_marks > 0
print("Total marks:", total_marks)
```

```python
#Step 2: Fix the Code

total_marks = 0

try:
    assert total_marks > 0, "Error: Total marks must be greater than zero."

    print("Total marks:", total_marks)

except AssertionError as e:  
    print(e)  
```
