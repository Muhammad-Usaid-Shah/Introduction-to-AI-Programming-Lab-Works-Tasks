LIST TASKS

```python
#1. Inventory System
#i. Add two new products
products = ["pencil", "notebook", "eraser", "marker"]
products.append("sharpener")
products.append("tape")
print(products)
```

```python
#ii. Insert product at position 2
products.insert(2, "highlighter")
print(products)
```

```python
#iii. Why list is appropriate?
## A list is appropriate because:
# - Items can change (add/remove/update)
# - Order matters
# - Inventory grows or shrinks
# - Lists allow fast indexing and modification
```

```python
#2. Data Slicing & Indexing
#i. First, last, second-to-last
temps = [23, 25, 19, 30, 21, 22]
print(temps[0])   # first
print(temps[-1])  # last
print(temps[-2])  # second-to-last
```

```python
#ii. Middle four values
print(temps[1:5])   # 25, 19, 30, 21
```

```python
#iii. Get “21” using negative indexing
print(temps[-2])
```

```python
#3. Nested Lists & Heterogeneous Data
#iii. Create 3 students
students = [
    ["Ali", 101, [80, 90, 85]],
    ["Sara", 102, [75, 88, 92]],
    ["John", 103, [60, 70, 65]]
]
print(students)
```

```python
#iii. Change one student’s second grade
students[1][2][1] = 95   
print(students)
```

```python
#iii. Add a fourth grade
students[0][2].append(88)   
print(students)
```

```python
#4. List Mutator Methods
colours = ["red", "green", "blue"]
```

```python
#i. append
colours.append("yellow")
```

```python
#ii. extend
colours.extend(["purple", "orange"])
```

```python
#iii. insert
colours.insert(1, "black")
```

```python
#iv. remove
colours.remove("green")
```

```python
#v. pop and print
removed = colours.pop()
print("Popped:", removed)
```

```python
#vi. Final list
print(colours)
```

```python
#5. Operators & Built-ins
numbers = [5, 10, 15, 20]
```

```python
#i. Check if 15 in list
print(15 in numbers)
```

```python
#ii. Concatenate
numbers = numbers + [25, 30]
print(numbers)
```

```python
#iii. Repeat list
print(numbers * 2)
```

```python
#iv. Built-ins
print(len(numbers))
print(min(numbers))
print(max(numbers))
print(sum(numbers))
```

```python
#6. List Comprehension & Conversion
values = (2, 4, 6, 8, 10)
```

```python
#i. Convert tuple → list
val_list = list(values)
print(val_list)
```

```python
#ii. Square each value
squared = [x*x for x in val_list]
print(squared)
```

```python
#iii. Filter > 30
filtered = [x for x in squared if x > 30]
print(filtered)
```

Tuple tasks

```python
#1. Device Configuration (Fixed)
device = ("device01", "192.168.1.10", 8080, True)
```

```python
#i. Access each by index
print(device[0])
print(device[1])
print(device[2])
print(device[3])
```

```python
#ii. Attempt to change port
try:
    device[2] = 9090
except TypeError as e:
    print("Error:", e)
```

```python
#iii. Convert → modify → back to tuple
dev_list = list(device)
dev_list[2] = 9090
device = tuple(dev_list)
print(device)
```

```python
#2. Creating Tuples & Single-Element Nuance
#i. Single element
single = (5,)
print(single)
```

```python
#ii. Not a tuple
not_tuple = (5)
print(type(not_tuple))   # int
```

```python
#iii. Tuple from range
t = tuple(range(5))
print(t)
```

```python
#3. Indexing, Slicing, Negative Indices
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")
```

```python
#i. First and last
print(weekdays[0])
print(weekdays[-1])
```

```python
#ii. Slice "Wed", "Thu", "Fri"
print(weekdays[2:5])
```

```python
#iii. Negative indexing “Sat”
print(weekdays[-2])
```

```python
#4. Nested Tuples & Mutable Elements
data = ("UserA", (2023, 9, 15), ["task1", "task2"])
```

```python
#i. Access date
date = data[1]
print(date[0], date[1], date[2])
```

```python
#ii. Modify list inside tuple
data[2].append("task3")
print(data)

# Explanation: tuple is immutable BUT it can contain mutable items (lists).
# The list inside tuple can be changed because the tuple stores a reference.
```

```python
#5. Packing & Unpacking
point = (10, 20, 30)
```

```python
#i. Unpack
x, y, z = point
print(x, y, z)
```

```python
#ii. Swap x and z
x, z = z, x
```

```python
#iii. Pack back
point2 = (x, y, z)
print(point)
print(point2)
```

```python
#6. Tuple vs List (Months)
#i. Define
months = ("Jan", "Feb", "Mar", "Apr", "May")
```

```python
#ii. Why tuple?
# Tuple is better because:
# - Months do NOT change
# - Fixed data → immutability is safer
```

```python
#iii. Try to append
try:
    months.append("NewMonth")
except AttributeError as e:
    print("Error:", e)
```

```python
#iv. Convert to list
months_list = list(months)
months_list.append("June")
print(months_list)
```

```python
#7. Tuple Operators & Built-ins
nums = (3, 6, 9, 12)
```

```python
#i. in / not in
print(6 in nums)
print(15 not in nums)
```

```python
#ii. Concatenate
print(nums + (15, 18))
```

```python
#iii. Multiply
print(nums * 3)
```

```python
#iv. Built-ins
print(len(nums))
print(min(nums))
print(max(nums))
print(sum(nums))
```
