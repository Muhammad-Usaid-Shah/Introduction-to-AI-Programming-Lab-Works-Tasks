While Loop

```python
#Task 1: Battery charging simulation

battery = 0
```

```python
while battery < 100:
    print("Charging...")
    battery += 10
```

```python
#Task 2: Countdown timer

timer = 10
```

```python
while timer >= 0:
    print("Test Running:", timer)
    timer -= 1
```

```python
#Task 3: Ask for voltage until user enters 0

voltage = input("Enter voltage (0 to stop): ")
```

```python
while voltage != "0":
    if voltage.isdigit():     # checks if input is a valid number
        print("You entered:", voltage)
    else:
        print("Invalid input! Please enter a number.")

    voltage = input("Enter voltage (0 to stop): ")
```

```python
#Task 4: Power doubling until it exceeds 1000 W

power = 2
```

```python
while power <= 1000:
    print("Power:", power)
    power *= 2
```

```python
#Task 5: Print "Cycle Running" until cycle reaches 5

cycle = 1
```
